```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 11. Categorize cars based on number of gears and count them
gear_counts = df['gear'].value_counts()
print("Number of cars in each gear category:\n", gear_counts)
```

```python
# 12. Relationship between fuel efficiency (mpg) and number of gears
plt.figure(figsize=(8, 5))
sns.boxplot(x='gear', y='mpg', data=df)
plt.xlabel("Number of Gears")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("Fuel Efficiency vs Number of Gears")
plt.show()
```

```python
df.corr()
```

```python
# 13. Relationship between horsepower and number of cylinders
plt.figure(figsize=(8, 5))
sns.scatterplot(x='hp', y='cyl', data=df)
plt.xlabel("Horsepower")
plt.ylabel("Number of Cylinders")
plt.title("Horsepower vs Number of Cylinders")
plt.show()
```

```python
# 14. Relationship between displacement and gross horsepower
plt.figure(figsize=(8, 5))
sns.scatterplot(x='disp', y='hp', data=df)
plt.xlabel("Displacement")
plt.ylabel("Gross Horsepower")
plt.title("Displacement vs Horsepower")
plt.show()
```

```python
# Data Analysis & Visualization
# 14. Explain the relationship between displacement and gross horse power
plt.scatter( df['hp'],df['disp'])
plt.title('Relationship between displacement and Horse Power (hp)  ')
plt.xlabel('horse power')
plt.ylabel('displacment')
plt.grid(True)
plt.show()
```
