Analyse and visualize the relationship between winning the toss and winning the match.

```python
df=pd.read_csv("matches.csv")
df
df['toss_match_win'] = df['toss_winner'] == df['winner']
toss_match_win_counts = df['toss_match_win'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=toss_match_win_counts.index, y=toss_match_win_counts.values, palette='Set2')
plt.xticks([0,1], ['Lost Match After Toss Win', 'Won Match After Toss Win'])
plt.title("Does Winning the Toss Help Win the Match?")
plt.ylabel("Number of Matches")
plt.tight_layout()
plt.show()

# Percentage\
win_pct = (df['toss_match_win'].sum() / len(df)) * 100
print(f"\nToss winners won the match in {win_pct:.2f}% of cases.")
print(toss_match_win_counts)
```

```python
import pandas as pd

df = pd.read_excel('cereals.xls', engine='xlrd')  # use the correct file path
print(df.head())  # shows the first 5 rows
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Hot vs Cold Cereals')
plt.ylabel('')
plt.show()

# Insight:
# Most cereals in the dataset are cold cereals.
```

```python
sns.boxplot(x='type', y='rating', data=df)
plt.title("Rating Distribution: Hot vs Cold Cereals")
plt.show()

# Insight:
# Cold cereals generally have a wider rating range and a slightly higher median.
```

```python
shelf_table = df['shelf'].value_counts().sort_index()
shelf_table.plot(kind='bar', title='Cereal Count by Shelf')
plt.xlabel("Shelf")
plt.ylabel("Count")
plt.show()

# Insight:
# Shelf 3 (highest shelf) has the most cereals, possibly more appealing to adults.
```

```python
sns.scatterplot(x='sugars', y='calories', data=df)
sns.regplot(x='sugars', y='calories', data=df, scatter=False, color='red')
plt.title("Sugar vs Calories")
plt.show()

# Insight:
# There is a moderate positive correlation â€” more sugar usually means more calories.
```

```python
sns.scatterplot(x='sugars', y='calories', data=df)
sns.regplot(x='sugars', y='calories', data=df)
plt.title("Sugar vs Calories")
plt.show()

# Insight:
# There is a moderate positive correlation â€” more sugar usually means more calories.
```

```python
threshold = df['rating'].median()
df['rating_category'] = df['rating'].apply(lambda x: 'High' if x > threshold else 'Low')
sns.boxplot(x='rating_category', y='sugars', data=df)
plt.title("Sugar in High vs Low Rated Cereals")
plt.show()

# Insight:
# High-rated cereals generally have less sugar compared to low-rated ones.
```

```python
df['healthy'] = (df['fiber'] >= 3) & (df['sugars'] <= 5)
sns.boxplot(x='healthy', y='rating', data=df)
plt.title("Rating: Healthy vs Non-Healthy Cereals")
plt.xticks([0, 1], ['Not Healthy', 'Healthy'])
plt.show()

# Insight:
# Healthy cereals tend to have better ratings on average.
```

```python
sns.boxplot(x='shelf', y='rating', data=df)
plt.title("Rating Distribution per Shelf")
plt.show()

# Insight:
# Shelf 3 cereals tend to have higher ratings â€” possibly more adult/healthy options.
```

```python
students = {
    "101": {"name": "Alice", "marks": [85, 90, 88]},
    "102": {"name": "Bob", "marks": [78, 82, 80]},
    "103": {"name": "Charlie", "marks": [92, 89, 95]},
    "104": {"name": "David", "marks": [70, 75, 72]},
    "105": {"name": "Eva", "marks": [88, 85, 90]}
}
```

```python
best_student = max(students.items(), key=lambda x: sum(x[1]['marks']))
print("Best-performing student:", best_student[1]['name'])
```

```python
subject_toppers = []
for i in range(3):  # 3 subjects
    highest = max(students.items(), key=lambda x: x[1]['marks'][i])
    subject_toppers.append(highest[1]['name'])

print("Subject-wise toppers:", subject_toppers)
```

```python
best_student = None
highest_total = 0

for student in students.values():
    total = sum(student['marks'])
    if total > highest_total:
        highest_total = total
        best_student = student['name']

print("Best-performing student:", best_student)
```

```python
subject_toppers = []

for i in range(3):  # loop for each subject
    highest_marks = 0
    topper = ""
    for student in students.values():
        if student['marks'][i] > highest_marks:
            highest_marks = student['marks'][i]
            topper = student['name']
    subject_toppers.append(topper)

print("Subject-wise toppers:", subject_toppers)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 5-number summary
summary = df['mpg'].describe()
print(summary[['min', '25%', '50%', '75%', 'max']])

# Boxplot to visualize outliers
sns.boxplot(x=df['mpg'])
plt.title("Boxplot of MPG")
plt.show()
```

```python
# 5-number summary
min_val = df['mpg'].min()
q1 = df['mpg'].quantile(0.25)
median = df['mpg'].median()
q3 = df['mpg'].quantile(0.75)
max_val = df['mpg'].max()

print("5-number summary of MPG:")
print("Min:", min_val)
print("Q1 (25%):", q1)
print("Median:", median)
print("Q3 (75%):", q3)
print("Max:", max_val)

# Outlier detection using IQR
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]

print("\nNumber of outliers in MPG column:", len(outliers))
if not outliers.empty:
    print("Outlier values:\n", outliers['mpg'].values)
else:
    print("No outliers detected.")
```
