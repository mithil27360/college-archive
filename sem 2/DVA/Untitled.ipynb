```python
import pandas as pd
import numpy as np

# Step 1: Generate a synthetic Cereals dataset
data = {
    "Cereal Name": ["Corn Flakes", "Choco Pops", "Oats", "Bran Flakes", "Wheat Crunch", 
                    "Nutri Bar", "Frosted Flakes", "Protein Cereal", "Honey Loops", "Granola"],
    "calories": [110, 120, 100, 90, 95, -1, 130, 140, 105, 115],  # -1 represents missing value
    "protein": [2, 3, -1, 4, 3, 2, 3, 5, 4, 3],  # -1 represents missing value
    "fat": [0.5, 1.0, -1, 0.8, 0.7, 0.4, 1.2, 0.5, 0.9, -1],  # Missing values
    "sodium": [200, 150, 180, 140, -1, 210, 160, 130, 190, 170],  # Missing value
    "fiber": [1.0, 0.5, 3.0, 2.5, 2.0, -1, 0.8, 4.0, 2.2, -1],  # Missing values
    "carbohydrates": [22.0, 24.0, 21.5, -1, 19.0, 25.0, 23.0, 20.5, 26.0, 24.5],  # Missing value
    "sugars": [2, 10, 1, 3, 4, 5, -1, 2, 9, 7],  # Missing value
    "potassium": [35, 40, 45, -1, 50, 30, 38, 42, 48, 37],  # Missing value
    "vitamins": [25, 30, 20, 35, -1, 15, 25, 10, 30, 40]  # Missing value
}

df = pd.DataFrame(data)
df.to_csv("Cereals.csv", index=False)
print("Cereals.csv created successfully!")

# # Automatically selects all numeric columns
cols_to_clean = df.select_dtypes(include=["number"]).columns 

# Step 1: Compute 5-number summary before  preprocessing
summary_before = df[cols_to_clean].describe().loc[["min", "25%", "50%", "75%", "max"]]
summary_before.to_csv("Summary_Before.csv")
print("5-number summary before preprocessing saved as Summary_Before.csv")

# Step 2: Handle Missing Data (-1) by replacing with mean
for col in cols_to_clean:
    mean_value = df[df[col] != -1][col].mean()  # Compute mean excluding -1
    df[col] = df[col].replace(-1, mean_value)

# Step 3: Compute 5-number summary after handling missing data
summary_after_missing = df[cols_to_clean].describe().loc[["min", "25%", "50%", "75%", "max"]]
summary_after_missing.to_csv("Summary_After_Missing.csv")
print("5-number summary after handling missing values saved as Summary_After_Missing.csv")

# Step 4: Handle Noisy Data by replacing outliers with median
for col in cols_to_clean:
    median_value = df[col].median()
    df[col] = df[col].where(df[col].between(df[col].quantile(0.05), df[col].quantile(0.95)), median_value)

# Step 5: Compute 5-number summary after handling noisy data
summary_after_noisy = df[cols_to_clean].describe().loc[["min", "25%", "50%", "75%", "max"]]
summary_after_noisy.to_csv("Summary_After_Noisy.csv")
print("5-number summary after handling noisy values saved as Summary_After_Noisy.csv")

# Step 6: Save cleaned dataset
df.to_csv("Cereals_Cleaned.csv", index=False)
print("Cleaned dataset saved as Cereals_Cleaned.csv")
```
