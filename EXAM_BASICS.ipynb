```python
nums = [11, -21, 0, 45, 66, -93]
pos = neg = 0

for i in nums:
    if i > 0:
        pos += 1
    elif i < 0:
        neg += 1

print("Positive:", pos)
print("Negative:", neg)
```

```python
# ðŸŒŸ 1. Updating a String
var1 = 'Hello World!'
# Replace part of the string using slicing and concatenation
print("Updated String:", var1[:6] + 'Python')  # Output: Hello Python

# ðŸŒŸ 2. String Formatting using % Operator
name = 'Abay'
weight = 55
# Use %s for string, %d for integer
print("My name is %s and weight is %d kg!" % (name, weight))  # Output: My name is Abay and weight is 55 kg!

# ðŸŒŸ 3. Built-in String Methods
str1 = "this is string example ... wow!!!"

# Capitalize the first letter
print("Capitalize:", str1.capitalize())  # Output: This is string example ... wow!!!

# Count occurrences of 's'
print("Count 's':", str1.count('s'))  # Output: 3

# Find position of substring
print("Find 'example':", str1.find('example'))  # Output: 15

# Convert all characters to lowercase
str2 = "THIS IS STRING EXAMPLE ... WOW!!!"
print("Lowercase:", str2.lower())  # Output: this is string example ... wow!!!

# Replace substring
str3 = "this is string example ... wow!!! this is really string"
print("Replace 'is' with 'was':", str3.replace("is", "was"))
# Output: thwas was string example ... wow!!! thwas was really string

# Swap the case of all letters
print("Swapcase:", str1.swapcase())  # Output: THIS IS STRING EXAMPLE ... WOW!!!

# Convert to title case (first letter of each word capitalized)
print("Title Case:", str1.title())  # Output: This Is String Example ... Wow!!!

# --------------------------------------------
# ðŸ“‹ 4. List Operations

lst = ['physics', 'chemistry', 1997, 2000]

# Append item to end of list
lst.append('maths')
print("After append:", lst)  # Output: ['physics', 'chemistry', 1997, 2000, 'maths']

# Insert at a specific index
lst.insert(2, 'biology')
print("After insert at index 2:", lst)
# Output: ['physics', 'chemistry', 'biology', 1997, 2000, 'maths']

```
