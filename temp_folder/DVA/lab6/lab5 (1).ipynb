 For the CEREALS dataset, perform data preprocessing and answer the following questions.
a. Create a table with the 5 number summary of all the numeric attributes. b. For each of the numeric attributes (proteins upto vitamins) , identify and replace all missing data(indicated with -1) with the arithmetic mean of the attribute. c. Create a table with the 5 number summary of all the numeric attributes after treating missing values. Do you think the strategy used in dealing with missing values was effective?
d. For each numeric attribute (proteins upto vitamins), identify and replace all noisy data with the median of attribute.
e. Create a table with the 5 number summary of all the numeric attributes after treating noisy values. Do you think the strategy used in dealing with noisy values was effective?

```python
df=pd.read_csv('Cereals.csv')
df
```

```python
numeric= df.select_dtypes(include=["number"]).columns 
numeric
```

```python
summary = df[numeric].describe().loc[["min", "25%", "50%", "75%", "max"]]
summary.to_csv("Summary.csv")
```

```python
for col in numeric:
    mean_value = df[df[col] != -1][col].mean()  # Compute mean excluding -1
    df[col] = df[col].replace(-1, mean_value)
```

```python
summary_new = df[ numeric].describe().loc[["min", "25%", "50%", "75%", "max"]]
summary_new.to_csv("Summary_After.csv")

```

```python
for col in numeric:
    median_value = df[col].median()
    df[col] = df[col].where(df[col].between(df[col].quantile(0.05), df[col].quantile(0.95)), median_value)
```

```python
summary_last = df[ numeric].describe().loc[["min", "25%", "50%", "75%", "max"]]
summary_last.to_csv("Summary_After_Noisy.csv")
print(" summary after handling noisy values ")

df.to_csv("Cereals_Cleaned.csv", index=False)

```
