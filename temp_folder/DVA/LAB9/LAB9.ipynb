```python
"""Tabulate and Visualize the top 5 countries that have posted the most texts with a 
positive sentiment. """
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

positive_posts = df[df['Sentiment'] == 'Positive']

country_counts = positive_posts['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Positive_Posts']


top_5_countries = country_counts.head(5)

print(top_5_countries)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_countries, x='Country', y='Positive_Posts', palette='viridis')
plt.xlabel("Country")
plt.ylabel("Number of Positive Posts")
plt.title("Top 5 Countries with the Most Positive Sentiment Posts")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for selected sentiments
filtered_df = df[df['Sentiment'].isin(['Positive', 'Negative', 'Neutral'])]

# Group by Year and Country, then count tweets
sentiment_counts = filtered_df.groupby(['Year', 'Country', 'Sentiment']).size().reset_index(name='Count')

# Pivot for visualization
pivot_table = sentiment_counts.pivot_table(index=['Year', 'Country'], columns='Sentiment', values='Count', fill_value=0)

# Print the table
print(pivot_table)

# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(data=sentiment_counts, x='Year', y='Count', hue='Sentiment')
plt.title('Yearly Sentiment Distribution Across Countries')
plt.xticks(rotation=45)
plt.show()

```
