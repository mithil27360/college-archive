"""Write a Python code block that inputs numbers into a list. Print
the largest, smallest, the sum, and the average of the numbers.
Count occurrences of a specific number in the list"""

```python
numbers = []
print("Enter numbers (type 'done' to finish):")
while True:
    user_input = input("Enter a number: ")
    if user_input.lower() == 'done':
        break
    try:
        number = float(user_input)  # Convert the input to a float
        numbers.append(number)
    except ValueError:
        print("Invalid input. Please enter a valid number.")

if numbers:
    largest = max(numbers)
    smallest = min(numbers)
    total_sum = sum(numbers)
    average = total_sum / len(numbers)
    
    print(f"\nLargest number: {largest}")
    print(f"Smallest number: {smallest}")
    print(f"Sum of numbers: {total_sum}")
    print(f"Average of numbers: {average}")
    
    search_number = float(input("\nEnter a number to count its occurrences: "))
    count = numbers.count(search_number)
    print(f"The number {search_number} occurs {count} times in the list.")
else:
    print("No numbers were entered.")
```

```python
# Input numbers into a list
numbers = []
print("Enter numbers (type 'done' to finish):")
while True:
    user_input = input("Enter a number: ")
    if user_input.lower() == 'done':
        break
    try:
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Invalid input. Please enter a valid number.")

if numbers:
    # Manual calculations
    largest = numbers[0]
    smallest = numbers[0]
    total_sum = 0
    count = 0

    for num in numbers:
        if num > largest:
            largest = num
        if num < smallest:
            smallest = num
        total_sum += num
        count += 1

    average = total_sum / count

    # Print results
    print("\nStatistics:")
    print(f"Largest number: {largest}")
    print(f"Smallest number: {smallest}")
    print(f"Sum of numbers: {total_sum}")
    print(f"Average of numbers: {average:.2f}")

    # Count occurrences of a specific number
    search_number = float(input("\nEnter a number to count its occurrences: "))
    occurrence_count = 0
    for num in numbers:
        if num == search_number:
            occurrence_count += 1
    print(f"The number {search_number} occurs {occurrence_count} times in the list.")
else:
    print("No numbers were entered.")
```

```python
# Dictionary of Cricket World Cup winners
world_cup_winners = {
    1975: "West Indies",
    1979: "West Indies",
    1983: "India",
    1987: "Australia",
    1992: "Pakistan",
    1996: "Sri Lanka",
    1999: "Australia",
    2003: "Australia",
    2007: "Australia",
    2011: "India",
    2015: "Australia",
    2019: "England"
}

# Manual counting
win_count = {}
unique_countries = []
for country in world_cup_winners.values():
    if country in win_count:
        win_count[country] += 1
    else:
        win_count[country] = 1
        unique_countries.append(country)

# Determine the best-performing country
best_country = None
max_wins = 0
for country, wins in win_count.items():
    if wins > max_wins:
        best_country = country
        max_wins = wins

# Display results
print(f"The best-performing country is: {best_country}")
print("Unique list of countries that have won the World Cup:")
print(", ".join(unique_countries))
```

```python
# Input a sentence from the user
sentence = input("Enter a sentence: ")

# Split the sentence into words
words = sentence.split()

# Initialize an empty dictionary to store word frequencies
word_frequency = {}

# Count the frequency of each word
for word in words:
    word = word.lower()  # Convert word to lowercase for case-insensitive comparison
    word_frequency[word] = word_frequency.get(word, 0) + 1

# Print the dictionary with word frequencies
print("Word frequencies:")
for word, frequency in word_frequency.items():
    print(f"{word}: {frequency}")
```

```python
"""
5. Write a Python code block to input numbers into two sets.
Perform union, intersection, and difference operations on the sets
and print the results"""
# Input two sets of numbers from the user using list comprehension
set1 = set(input("Enter numbers for the first set (separated by space): ").split())
set2 = set(input("Enter numbers for the second set (separated by space): ").split())

# Perform union operation
union_result = set1 | set2  # Using the '|' operator for union
print("Union of set1 and set2:", union_result)

# Perform intersection operation
intersection_result = set1 & set2  # Using the '&' operator for intersection
print("Intersection of set1 and set2:", intersection_result)

# Perform difference operation (set1 - set2)
difference_result = set1 - set2  # Using the '-' operator for difference
print("Difference of set1 and set2 (set1 - set2):", difference_result)

# Perform difference operation (set2 - set1)
difference_result_reverse = set2 - set1  # Using the '-' operator for difference
print("Difference of set2 and set1 (set2 - set1):", difference_result_reverse)
```

```python
# Input two sets of numbers from the user
set1 = set(map(int, input("Enter numbers for the first set (separated by space): ").split()))
set2 = set(map(int, input("Enter numbers for the second set (separated by space): ").split()))

# Perform union operation
union_result = set1.union(set2)
print("Union of set1 and set2:", union_result)

# Perform intersection operation
intersection_result = set1.intersection(set2)
print("Intersection of set1 and set2:", intersection_result)

# Perform difference operation (set1 - set2)
difference_result = set1.difference(set2)
print("Difference of set1 and set2 (set1 - set2):", difference_result)

# Perform difference operation (set2 - set1)
difference_result_reverse = set2.difference(set1)
print("Difference of set2 and set1 (set2 - set1):", difference_result_reverse)
```
