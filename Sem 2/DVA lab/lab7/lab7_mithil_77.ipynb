# 2. Visualize the total number of matches won by each team.

# 9. List and visualize the best teams(top 5), season-wise.

```python
season_winners = df.groupby(['season', 'winner']).size().reset_index(name='wins')
top_teams = season_winners.groupby('season').apply(lambda x: x.nlargest(5, 'wins'))
plt.figure(figsize=(14,6))
sns.barplot(data=top_teams, x='season', y='wins', hue='winner')
plt.xlabel("Season")
plt.ylabel("Wins")
plt.title("Top 5 Teams Season-wise")
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
df['toss_win_match_win'] = df['toss_winner'] == df['winner']
toss_impact = df['toss_win_match_win'].value_counts(normalize=True) * 100
plt.figure(figsize=(8,6))
sns.barplot(x=toss_impact.index, y=toss_impact.values, palette=['red', 'blue'])
plt.xticks([0, 1], ['Lost After Toss Win', 'Won After Toss Win'])
plt.ylabel("Percentage")
plt.title("Impact of Toss on Match Outcome")
plt.show()
```
