```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the datasets
digimon_df = pd.read_csv("DigiDB_digimonlist.csv")  
move_df = pd.read_csv("DigiDB_movelist.csv")
support_df = pd.read_csv("DigiDB_supportlist.csv")
```

```python

print("Digimon List:")
print(digimon_df.head(), "\n")

print("Move List:")
print(move_df.head(), "\n")

print("Support List:")
print(support_df.head(), "\n")
```

```python
# Checking the structure of each dataset
print("Digimon List Info:")
print(digimon_df.info(), "\n")

print("Move List Info:")
print(move_df.info(), "\n")

print("Support List Info:")
print(support_df.info(), "\n")
```

```python
# Summary statistics for numerical columns
print("Digimon List Summary:")
print(digimon_df.describe(), "\n")

print("Move List Summary:")
print(move_df.describe(), "\n")

print("Support List Summary:")
print(support_df.describe(), "\n")
```

```python
# Value counts for Digimon Types (sorted)
print("Value Counts for Digimon Types:")
print(digimon_df['Digimon'].value_counts().sort_values(ascending=False), "\n")

# Value counts for Move Types (sorted)
print("Value Counts for Move Types:")
print(move_df['Move'].value_counts().sort_values(ascending=False), "\n")
```

```python
# Value counts for Digimon Types (sorted)
print("Value Counts for Digimon Stage:")
print(digimon_df['Stage'].value_counts().sort_values(ascending=False), "\n")

print("Value Counts for Digimon Types:")
print(digimon_df['Type'].value_counts().sort_values(ascending=False), "\n")

print("Value Counts for Digimon Attribute:")
print(digimon_df['Attribute'].value_counts().sort_values(ascending=False), "\n")

print("Value Counts for Digimon Memory:")
print(digimon_df['Memory'].value_counts().sort_values(ascending=False), "\n")

print("Value Counts for Digimon Equip Slots:")
print(digimon_df['Equip Slots'].value_counts().sort_values(ascending=False), "\n")

```
