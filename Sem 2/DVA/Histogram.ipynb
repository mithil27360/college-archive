```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('mtcars.csv')

# Select numeric columns
numeric_cols = df.select_dtypes(include='number').columns

# Set the number of rows and columns for the subplots
n_cols = 3
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols  # ceiling division

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
axes = axes.flatten()  # Flatten 2D axes array to 1D for easy indexing

# Plot each histogram
for i, col in enumerate(numeric_cols):
    sns.histplot(df[col], kde=True, ax=axes[i], bins=10, color='skyblue')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
cylinders = df['cyl'].unique()
data = [  df[ df['cyl'] == cyl ] ['mpg'] for cyl in cylinders  ]

plt.boxplot(data, labels=cylinders)
plt.title('MPG by Cylinder Count')
plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
sns.boxplot(y='mpg', x='cyl', hue='am', data=df, palette='coolwarm')
plt.title('MPG by Cylinders and Transmission (am)')
plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.legend(title='Transmission (0=Auto, 1=Manual)')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("mtcars.csv")

# KDE plot for a single numeric column (e.g., mpg)
sns.kdeplot(df['mpg'], shade=True, color='skyblue')
plt.title('KDE Plot of MPG')
plt.xlabel('Miles per Gallon')
plt.ylabel('Density')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure()
    sns.kdeplot(df[col], shade=True, color='orange')
    plt.title(f'KDE Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
# Compare 'mpg' distribution by 'am' (transmission: 0 = auto, 1 = manual)
sns.kdeplot(data=df[df['am'] == 0]['mpg'], shade=True, label='Automatic', color='red')
sns.kdeplot(data=df[df['am'] == 1]['mpg'], shade=True, label='Manual', color='green')
plt.title('KDE of MPG by Transmission')
plt.xlabel('MPG')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mtcars.csv")

# Count of each cylinder category
cyl_counts = df['cyl'].value_counts()

# Plot bar chart
plt.bar(cyl_counts.index, cyl_counts.values, color='skyblue')
plt.title('Count of Cars by Cylinder')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(y='cyl', data=df, palette='Set3')
plt.title('Cars by Cylinder (Horizontal)')
plt.xlabel('Count')
plt.ylabel('Number of Cylinders')
plt.show()

```
