```python
# Assistant
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Set style for visualizations
plt.style.use('ggplot')
sns.set(style="whitegrid")

# Data Exploration of digimon_df
print("DIGIMON DATASET EXPLORATION")
print("-" * 50)

# Basic information
print(f"Shape of digimon dataset: {digimon_df.shape}")
print(f"Number of unique Digimon: {digimon_df['Digimon'].nunique()}")
print("\nFirst 5 rows:")
print(digimon_df.head())

# Summary statistics
print("\nSummary statistics for numerical columns:")
print(digimon_df.describe())

# Check for missing values
print("\nMissing values in digimon dataset:")
print(digimon_df.isnull().sum())

# Distribution of categorical variables
print("\nDistribution of Digimon by Stage:")
print(digimon_df['Stage'].value_counts())

print("\nDistribution of Digimon by Type:")
print(digimon_df['Type'].value_counts())

print("\nDistribution of Digimon by Attribute:")
print(digimon_df['Attribute'].value_counts())

# Visualize some distributions
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(digimon_df['Memory'], kde=True)
plt.title('Distribution of Memory')

plt.subplot(1, 2, 2)
sns.histplot(digimon_df['Lv 50 HP'], kde=True)
plt.title('Distribution of HP at Level 50')

plt.tight_layout()
plt.show()

# Data Exploration of move_df
print("\n\nMOVE DATASET EXPLORATION")
print("-" * 50)

# Basic information
print(f"Shape of move dataset: {move_df.shape}")
print(f"Number of unique Moves: {move_df['Move'].nunique()}")
print("\nFirst 5 rows:")
print(move_df.head())

# Summary statistics
print("\nSummary statistics for numerical columns:")
print(move_df.describe())

# Check for missing values
print("\nMissing values in move dataset:")
print(move_df.isnull().sum())

# Distribution of categorical variables
print("\nDistribution of Moves by Type:")
print(move_df['Type'].value_counts())

print("\nDistribution of Moves by Attribute:")
print(move_df['Attribute'].value_counts())

print("\nDistribution of Inheritable Moves:")
print(move_df['Inheritable'].value_counts())

# Visualize some distributions
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(move_df['SP Cost'], kde=True)
plt.title('Distribution of SP Cost')

plt.subplot(1, 2, 2)
sns.histplot(move_df['Power'], kde=True)
plt.title('Distribution of Move Power')

plt.tight_layout()
plt.show()

# Data Exploration of support_df
print("\n\nSUPPORT DATASET EXPLORATION")
print("-" * 50)

# Basic information
print(f"Shape of support dataset: {support_df.shape}")
print(f"Number of unique Support Skills: {support_df['Support'].nunique()}")
print("\nFirst 5 rows:")
print(support_df.head())

# Check for missing values
print("\nMissing values in support dataset:")
print(support_df.isnull().sum())

# Word count in descriptions
support_df['Description_Length'] = support_df['Description'].str.len()

plt.figure(figsize=(10, 5))
sns.histplot(support_df['Description_Length'], kde=True)
plt.title('Distribution of Support Skill Description Length')
plt.xlabel('Character Count')
plt.show()

# Relationships between datasets
print("\n\nRELATIONSHIPS BETWEEN DATASETS")
print("-" * 50)

# Check if all Digimon have a corresponding Support skill
digimon_support_match = digimon_df['Digimon'].isin(support_df['Support'])
print(f"Percentage of Digimon with matching Support skill: {digimon_support_match.mean()*100:.2f}%")

# Summary of the datasets
print("\nDATASET SUMMARY")
print("-" * 50)
print(f"1. Digimon Dataset: {digimon_df.shape[0]} Digimon with {digimon_df.shape[1]} attributes")
print(f"2. Move Dataset: {move_df.shape[0]} moves with {move_df.shape[1]} attributes")
print(f"3. Support Dataset: {support_df.shape[0]} support skills with {support_df.shape[1]} attributes")
```
