```python
 total_runs = df.groupby('batting_team')['total_runs'].sum().reset_index()
print(total_runs)
```

```python
# Group by 'batting_team' and calculate the total runs
total_runs = df.groupby('batting_team')['total_runs'].sum()


# Create a bar plot to visualize total runs by each team
plt.figure(figsize=(10, 6))

total_runs.plot(kind='bar')


plt.title('Total Runs Scored by Each Team')
plt.xlabel('Total Runs')
plt.ylabel('Teams')



plt.xticks(rotation=45,ha='right')

plt.show()
```

Identify and display the names of the top 10 highest-scoring batters with their total runs scored.

```python
top_batter=df.groupby('batter')['batsman_runs'].sum()
top_sort=top_batter.sort_values(ascending=False)
print(top_sort.head(10))

plt.figure(figsize=(8,6))
top_sort.head(10).plot(kind='bar')
plt.title('top batsmen')
plt.xlabel("batsmen")
plt.ylabel("runs")
plt.show()
```

 Find and visualize the top 10 bowlers with the most wickets. 

```python
top_bowl=df.groupby('bowler')['is_wicket'].sum()
top_sort=top_bowl.sort_values(ascending=False)
print(top_sort.head(10))

plt.figure(figsize=(8,6))
top_sort.head(10).plot(kind='bar')
plt.title('top wicket takers')
plt.xlabel("bowlers")
plt.ylabel("wickets")
plt.show()
```

```python
#5. Tabulate and visualize the batsmen (top 5) involved in the most run-outs.

run_outs = df[df['dismissal_kind'] == 'run out']


top_run_outs = run_outs['batter'].value_counts().head(5)


print(top_run_outs)


plt.figure(figsize=(10,5))
top_run_outs.plot(kind='bar', color='coral')
plt.xlabel('Batsman')
plt.ylabel('Number of Run Outs')
plt.title('Top 5 Batsmen Involved in the Most Run Outs')
plt.xticks(rotation=45)
plt.show()
```

```python
x = df.groupby('over')['total_runs'].sum() / 6  
top_3_overs = x.sort_values(ascending=False).head(3)

print(top_3_overs)

plt.figure(figsize=(8,6))
top_3_overs.plot(kind='bar', color='blue')
plt.xlabel("Over")
plt.ylabel("Run Rate")
plt.title("Top 3 Overs with the Highest Run Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by team and sum runs for the first 6 overs (Powerplay)
powerplay_runs = df[df['over'] <= 6].groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(12, 6))
powerplay_runs.plot(kind='bar', color='dodgerblue', edgecolor='black')

plt.xlabel("Team", fontsize=14)
plt.ylabel("Total Runs", fontsize=14)
plt.title("Total Runs in Powerplay (First 6 Overs)", fontsize=16, fontweight='bold')

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)


plt.show()
```

```python
###Tabulate and visualize the top 5 batsmen with the most sixes and fours.
import pandas as pd
import matplotlib.pyplot as plt

# Filter only boundaries
sixes = df[df['batsman_runs'] == 6].groupby('batter').size().sort_values(ascending=False).head(5)
fours = df[df['batsman_runs'] == 4].groupby('batter').size().sort_values(ascending=False).head(5)

# Combine the data
boundaries = pd.DataFrame({'Fours': fours, 'Sixes': sixes}).fillna(0)

# Plot
boundaries.plot(kind='bar', figsize=(10,6), color=['blue', 'red'], edgecolor='black')

plt.xlabel("Batsman", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Top 5 Batsmen with the Most Fours & Sixes", fontsize=16, fontweight='bold')

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(["Fours", "Sixes"])

plt.show()
```
