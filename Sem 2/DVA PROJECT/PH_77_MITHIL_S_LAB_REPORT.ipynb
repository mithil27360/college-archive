## üîó Dataset Integration

---

To unify the datasets for analysis:

- **We merge ** using common fields such as `Attribute` .
- Resolved mismatched attribute names by standardizing values (e.g., replacing `"Thunder"` with `"Electric"`).
- Ensured consistency across all attribute columns to allow accurate grouping and analysis.

---


```python
# üîπ Step 1: Clean and standardize the 'Attribute' column
digimon_df["Attribute"] = digimon_df["Attribute"].str.strip().str.title()
move_df["Attribute"] = move_df["Attribute"].str.strip().str.title()

# üîÑ Step 2: Replace 'Thunder' with 'Electric'
move_df["Attribute"] = move_df["Attribute"].replace("Thunder", "Electric")

# ‚úÖ Step 3: Merge on 'Attribute' only (inner = only matching attributes)
merged_df = pd.merge(digimon_df, move_df, on="Attribute", how="inner")

# üíæ Step 4: Save the merged result
merged_df.to_csv("merged_digimon_clean.csv", index=False)

# üîç Preview
print("Merged dataset shape:", merged_df.shape)
merged_df.head()
```

## üîó Resolving Duplicate Merge Entries

---

To avoid repeated rows during merge (e.g., multiple "Kuramon" entries for each matching move), we selected only the **highest Power move per Attribute** using `groupby`. This ensures that each Digimon is paired with **one optimal move**, eliminating redundancy and improving interpretability of the merged dataset.

---


```python
#
# Standardize attribute column
digimon_df["Attribute"] = digimon_df["Attribute"].str.strip().str.title()
move_df["Attribute"] = move_df["Attribute"].str.strip().str.title()

# Replace Thunder with Electric
move_df["Attribute"] = move_df["Attribute"].replace("Thunder", "Electric")

# Convert Power to numeric, errors='coerce' makes sure non-numeric values become NaN
move_df["Power"] = pd.to_numeric(move_df["Power"], errors='coerce')

# üîç Keep only the move with highest Power per Attribute
best_moves = move_df.sort_values("Power", ascending=False).dropna(subset=["Power"])
best_moves = best_moves.groupby("Attribute").first().reset_index()

# ‚úÖ Now merge with digimon_df on Attribute (1 move per Attribute only)
merged_df = pd.merge(digimon_df, best_moves, on="Attribute", how="inner")

# üíæ Save the cleaned and reduced merged dataset
merged_df.to_csv("merged_clean_summary.csv", index=False)

# Optional: Display first few rows
print("Final merged dataset shape:", merged_df.shape)
print(merged_df.head())
```

## üßπ Data  Integration Summary

---

- Standardized the `Attribute` column across datasets (e.g., `"thunder"` ‚Üí `"Electric"`).
- Converted the `Power` column in the move dataset to numeric for analysis.
- Selected the **highest Power move per Attribute** to represent the most impactful attack.
- Performed an **inner merge** on the `Attribute` column to combine Digimon stats with their strongest matching moves.
- Saved the final cleaned dataset for further analysis.

## üìä Digimon with Maximum and Minimum Stats at Level 50

Created a horizontal bar plot to visually compare the **highest** and **lowest** performing Digimon for each stat.  
Each bar is labeled with the Digimon name and attribute, making it easy to interpret extremes across stats like HP, SP, Attack, etc.


```python

df = pd.read_csv("DigiDB_digimonlist.csv")
df.columns = df.columns.str.replace(" ", "")

# Define stats
stats = ["Lv50HP", "Lv50SP", "Lv50Atk", "Lv50Def", "Lv50Int", "Lv50Spd"]

# Prepare data for plotting
data = []

for stat in stats:
    max_row = df.loc[df[stat].idxmax()]
    min_row = df.loc[df[stat].idxmin()]
    
    data.append({
        "Stat": stat,
        "Value": max_row[stat],
        "Digimon": max_row["Digimon"],
        "Attribute": max_row["Attribute"],
        "Type": "Highest"
    })
    
    data.append({
        "Stat": stat,
        "Value": min_row[stat],
        "Digimon": min_row["Digimon"],
        "Attribute": min_row["Attribute"],
        "Type": "Lowest"
    })

plot_df = pd.DataFrame(data)

# Combine Digimon name + Attribute
plot_df["Label"] = plot_df["Digimon"] + " (" + plot_df["Attribute"] + ")"

# Set style
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# Sort for clean horizontal plotting
plot_df["StatType"] = plot_df["Stat"] + " - " + plot_df["Type"]
plot_df = plot_df.sort_values("Value", ascending=True)

# Create barplot
barplot = sns.barplot(data=plot_df, x="Value", y="StatType", hue="Type", palette=["#4CAF50", "#F44336"])

# Add text labels on the bars
for index, row in plot_df.iterrows():
    barplot.text(row["Value"] + 5, 
                 index, 
                 f"{row['Label']} ({int(row['Value'])})", 
                 color='black', 
                 va="center", 
                 fontsize=9)

# Final touches
plt.title("Digimon with Maximum and Minimum Stats at Level 50", fontsize=16)
plt.xlabel("Stat Value")
plt.ylabel("Stat - Type")
plt.legend(title="Stat Type", loc='lower right')
plt.tight_layout()
plt.show()
```
