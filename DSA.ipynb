## DICE
You throw a dice n times, and every throw produces an outcome between 1 and 6. What is the probability that the sum of outcomes is between a and b?
Input
The only input line contains three integers n, a and b.
Output
Print the probability rounded to six decimal places (rounding half to even).
Constraints

1 \le n \le 100
1 \le a \le b \le 6n

Example
Input:
2 9 10

Output:
0.194444

```python
import random

def dice():
     return random.randint(1, 6)


n=int(input("enter n"))


sum=0


for n in range(1,100):
    a= dice()
    sum=sum+a
print(sum)
    
    


    
```

```python
# import random

def dice():
    return random.randint(1, 6)

n = int(input("Enter number of dice rolls (n): "))
a = int(input("Enter lower bound (a): "))
b = int(input("Enter upper bound (b): "))

# Number of simulations to estimate probability
trials = 10000000
success = 0

for _ in range(trials):
    total = 0
    for _ in range(n):
        total += dice()
    if a <= total <= b:
        success += 1

probability = success / trials
print(f"{probability:.6f}")
```
