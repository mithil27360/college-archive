Write a Python code block that inputs numbers into a list. Print 
the largest, smallest, the sum, and the average of the numbers. 
Count occurrences of a specific number in the list

```python
numbers=[]
n=int(input("enter number of digits  in the list: "))
for num in range(0,n):
    number=int(input("enter a number "))
    numbers.append(number)
maximum=max(numbers)
minimum=min(numbers)
sum1=sum(numbers)
digits=len(numbers)
average=sum1/digits
print(f"maximum is {maximum}")
print(f"minimum is {minimum}")
print(f"sum is {sum1}")
print(f"average is {average}")
search=int(input("enter number to search : "))
count=numbers.count(search)
print(f"{search} occured {count} times in the list ")
```

Write a Python code block to create a dictionary of cricket World 
Cup winners. Let the key be the year; the value is the country that 
won the World Cup that year.### Print the name of the best
performing country.  Display the unique list of countries that 
have won the World Cup.

```python

winners = {
    1983: "India",
    1987: "Australia",
    1992: "Pakistan",
    1996: "Sri Lanka",
    1999: "Australia",
    2003: "Australia",
    2007: "Australia",
    2011: "India",
    2015: "Australia",
    2019: "England",
    2023: "Australia"
}

country_count = {}
for country in winners.values():
    if country in country_count:
        country_count[country] += 1
    else:
        country_count[country] = 1

best = max(country_count, key=country_count.get)
count = country_count[best]
print(f"Best-performing country: {best} ({count} wins)")

unique = set(winners.values())
print("Unique World Cup-winning countries:", unique)
```

 Write a Python code block that inputs a sentence from the user. 
Count the frequency of each word in the sentence and store the 
result in a dictionary. Prints the dictionary with words as keys 
and their frequencies as values.

```python
sentence = input("Enter a sentence: ").lower().split()
word_count = {}

for word in sentence:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print("Word Frequency Dictionary:", word_count)
```

Write a Python code block to input numbers into two sets.  
Perform union, intersection, and difference operations on the sets 
and print the results. 

```python
set1=set(input("enter set1  :  "))
set2=set(input("enter set 2  :  "))
print(f"set1 is {set1}")
print(f"set2 is {set2}")
union=set1|set2
intersection=set1 & set2
difference=set1-set2
rev_diff=set2-set1
print("union  :  ",union)
print("intersection  :  ",intersection)
print("difference  :  ",difference)
print("rev_diff  :  ",rev_diff)
```
