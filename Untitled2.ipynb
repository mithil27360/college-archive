```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is the DataFrame and 'mpg', 'cylinders', 'weight' are column names
# Example: df = pd.read_csv('auto-mpg.csv')

# A. Display a Histogram of the MPG column
plt.figure(figsize=(8, 5))
plt.hist(df['mpg'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# B. Tabulate and visualize the distribution of MPG for each Cylinder Category
# Tabulation
mpg_by_cylinder = df.groupby('cyl')['mpg'].describe()
print("MPG Distribution by Cylinders:\n", mpg_by_cylinder)

# Visualization using Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='cyl', y='mpg', palette='Set2')
plt.title('MPG Distribution by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.grid(True)
plt.show()

# C. Tabulate and visualize the relationship between MPG and weight of the car
# Tabulation: correlation table
correlation = df[['mpg', 'wt']].corr()
print("Correlation between MPG and Weight:\n", correlation)

# Visualization using Scatterplot with regression line
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='wt', y='mpg', color='purple')
sns.regplot(data=df, x='wt', y='mpg', scatter=False, color='black')  # regression line
plt.title('Relationship between MPG and Car Weight')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.grid(True)
plt.show()
```
