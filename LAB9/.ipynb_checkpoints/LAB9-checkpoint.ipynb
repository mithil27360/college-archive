# Find and display the text(s) with the maximum number of likes

```python


max_likes = df["Likes"].max()

# Find the text(s) with the maximum number of likes
most_liked_texts = df[df["Likes"] == max_likes][["Text", "Likes"]]

# Display the result
print(most_liked_texts)
```

```python
max_REPO = df["Retweets"].max()

# Find the text(s) with the maximum number of likes
most_REPO = df[df["Retweets"] == max_REPO][["Text", "Retweets",'Sentiment']]

# Display the result
print(most_REPO)
```
