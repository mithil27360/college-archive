```python

# 8. Tabulate and visualize cereals by display shelf
shelf_distribution = df["shelf"].value_counts().sort_index()
plt.figure(figsize=(6, 4))
sns.barplot(x=shelf_distribution.index, y=shelf_distribution.values, palette="viridis")
plt.xlabel("Display Shelf")
plt.ylabel("Number of Cereals")
plt.title("8. Cereal Count by Display Shelf")
plt.show()
```

```python
# 9. How does the sugar content vary across different brands?
plt.figure(figsize=(10, 5))
sns.boxplot(x="mfr", y="sugars", data=df, palette="Set2")
plt.xlabel("Manufacturer")
plt.ylabel("Sugar Content")
plt.title("9. Sugar Content Variation Across Brands")
plt.show()
```

```python
# 10. What is the average calories per manufacturer?
avg_calories_per_mfr = df.groupby("mfr")["calories"].mean()
print("\n10. Average Calories per Manufacturer:\n", avg_calories_per_mfr)

# 11. Average nutritional content across all cereals
avg_nutrition = df[["calories", "sugars", "protein", "fiber", "fat", "sodium", "carbo", "potass"]].mean()
print("\n11. Average Nutritional Content Across All Cereals:\n", avg_nutrition)
```

```python
# 12. Relationship between sugar and calories
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df["sugars"], y=df["calories"], alpha=0.7)
plt.xlabel("Sugar (g)")
plt.ylabel("Calories")
plt.title("12. Relationship Between Sugar and Calories")
plt.show()
```

```python
# 13. How does sugar content compare in high vs. low ratings?
df["high_rating"] = df["rating"] >= df["rating"].median()
plt.figure(figsize=(6, 4))
sns.boxplot(x="high_rating", y="sugars", data=df, palette="coolwarm")
plt.xlabel("High Rating (True/False)")
plt.ylabel("Sugar Content")
plt.title("13. Sugar Content in High vs. Low Rated Cereals")
plt.show()
```

```python
# 14. Do healthy cereals (high fiber, low sugar) have higher ratings?
healthy_cereals = df[(df["fiber"] > 3) & (df["sugars"] < 5)]
average_rating_healthy = healthy_cereals["rating"].mean()
average_rating_all = df["rating"].mean()
print(f"\n14. Average Rating for Healthy Cereals: {average_rating_healthy}")
print(f"    Average Rating for All Cereals: {average_rating_all}")

# 15. Relationship between rating and display shelf
plt.figure(figsize=(6, 4))
sns.boxplot(x="shelf", y="rating", data=df, palette="coolwarm")
plt.xlabel("Display Shelf")
plt.ylabel("Rating")
plt.title("15. Relationship Between Rating and Display Shelf")
plt.show()
```
